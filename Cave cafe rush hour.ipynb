## Rush hour analysis

```python
print(df['StartSession'].dtype)
```

```python
df['Month']=df['StartSession'].dt.month
df['week']=df['StartSession'].dt.isocalendar().week
df['day']=df['StartSession'].dt.day_name()
df['hour']=df['StartSession'].dt.hour
```

```python
df.head()
```

```python
def get_shift(hour):
    if 6 <= hour< 18:
        return 'Day Shift'
    else:
        return 'Night Shift'
df['Shift'] = df['hour'].apply(get_shift)
```

```python
df.head()
```

## Plotting the Rush hour

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='hour', hue='Shift')
plt.title('Rush Hour Analysis by Shift')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Sessions')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='day', hue='Shift', order=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
plt.title('Rush days Analysis by Shift')
plt.xlabel('Days of the Week')
plt.ylabel('Number of Sessions')
plt.show()
```

```python
df = df.sort_values(by='StartSession')
df = df.sort_index()
```

```python
df['StartSession']= pd.to_datetime(df['StartSession'])
df = df.set_index('StartSession')
session_per_day = df.resample('D').size()
```

```python
plt.figure(figsize=(14,6))
session_per_day.plot()
plt.title('Number of Sessions per Day')
plt.xlabel('Date')
plt.ylabel('Number of Sessions')
plt.grid(True)
plt.show()
```

```python
top_months = df.groupby(['Month', 'hour']).size().reset_index(name='SessionCount')
top_months = top_months.sort_values(by='SessionCount', ascending=False)
```

```python
top_months.head(10)
```

```python
pivot = top_months.pivot(index='hour', columns='Month', values='SessionCount')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Rush Hour Sessions by Month and Hour')
plt.xlabel('Month')
plt.ylabel('Hour of Day')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=top_months, x='hour', y='SessionCount', hue='Month')
plt.title('Rush Hour Sessions by Hour and Month')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Sessions')
plt.legend(title='Month')
plt.show()
```
